```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_data(url):
    response = requests.get(url)
    if response.status_code != 200:
        raise Exception(f"Failed to load page: {url}")

    soup = BeautifulSoup(response.text, 'html.parser')
    
    enrollment = soup.find('li', attrs={'data-grafico': '5908'}).find('p', class_='ind-value').get_text(strip=True).split('e')[0]
    hdi = soup.find('li', attrs={'data-grafico': '30255'}).find('p', class_='ind-value').get_text(strip=True).split('[')[0]
    income = soup.find('li', attrs={'data-grafico': '48986'}).find('p', class_='ind-value').get_text(strip=True).split('R')[0]

    return enrollment, hdi, income

url = "https://www.ibge.gov.br/en/cities-and-states/sp.html"
try:
    enrollment, hdi, income = scrape_data(url)
    print(enrollment, hdi, income)
except Exception as e:
    print(str(e))
```

```python
states = ["ac", "al", "ap", "am", "ba", "ce", "df", "es", "go", "ma", "mt", "ms", "mg", "pa", "pb", "pr", "pe", "pi",
          "rj", "rn", "rs", "ro", "rr", "sc", "sp", "se", "to"]

base_url = "https://www.ibge.gov.br/en/cities-and-states/{}.html"
```

```python
import pandas as pd

data = pd.DataFrame(columns=['State', 'Enrollment', 'HDI', 'Income'])

for state in states:
    url = base_url.format(state)
    enrollment, hdi, income = scrape_data(url)
    new_r = pd.DataFrame({'State':state, 'Enrollment': enrollment, 'HDI': hdi, 'Income':income}, index=[0])
    data = pd.concat([data, new_r], ignore_index=True)
data
```

```python
data.to_csv('Brazil_macroecon.csv')
```
